#### Data Cleaning: Here we will assess the dataset's quality, that is; we will check for missing values, the presence of outliers, duplicates,  and handle them accordingly

```python
# Checking  for missing values in the entire dataset
spotify.isnull()
```

```python
#Check for sum missing values
spotify.isnull().sum()
```

```python
# Separating numeric and non-numeric columns
numeric_columns= spotify.select_dtypes(include =['number']).columns
non_numeric_columns = spotify.select_dtypes(include=['number']).columns

#Printing the numeric and non_numeric columns
print("numeric_data:",numeric_columns)
print("non_numeric_data", non_numeric_columns)
```

```python
# Filling in the missing values in numeric columns with mode and non-numeric columns with mode(the most common value)
spotify[numeric_columns] = spotify[numeric_columns].fillna(spotify[numeric_columns]).mode().iloc[0]
spotify[non_numeric_columns] = spotify[non_numeric_columns].fillna(spotify[non_numeric_columns].mode().iloc[0])

# Checking for missing values in the entire dataset
print(spotify.isnull().sum())
```

##### *Since the non-numeric column key has still not been filled we will repeat the process but be specific that we are filling the missing value in the key column*

```python
# Filling the missing values in the key column with the mode
spotify['key'] = spotify['key'].fillna(spotify['key'].mode().iloc[0])

print(spotify.isnull().sum())
```

##### *From the above key we can see that the missing values in the key column have been filled with the mode*

```python
# Checking for duplicates in the dataset
duplicate_rows = spotify.duplicated

# Print the number of duplicate rows
print(spotify.duplicated())

# Rather display the duplicated rows
print(spotify[spotify.duplicated()])
```

```python
# Drop the duplicates
spotify = spotify.drop_duplicates() 

# Checking for duplicates in the dataset
duplicate_rows = spotify.duplicated

print(spotify.duplicated().sum())
```

```python
# Checking for the presence of outliers in the dataset
# Function to detect outliers using IQR for all numeric columns
def detect_outliers_iqr(data):
    outliers_dict = {}  # Dictionary to store outliers information for each column
    for column in data.select_dtypes(include=['number']).columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers
        outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]

        # Store the outliers count in the dictionary
        outliers_dict[column] = len(outliers)

    return outliers_dict

# Apply the function and print the number of outliers per numeric column
outliers_summary = detect_outliers_iqr(spotify)

# Display the summary of outliers
for column, outliers_count in outliers_summary.items():
    print(f"{column} has {outliers_count} outliers")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to plot all boxplots in one chart with subplots
def plot_boxplots_in_grid(data):
    numeric_columns = data.select_dtypes(include=['number']).columns
    num_columns = len(numeric_columns)
    
    # Create a grid of subplots (determine rows and columns based on number of numeric columns)
    cols = 3  
    rows = (num_columns // cols) + (num_columns % cols > 0)  
    
    # Create a figure with subplots
    fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15, 6 * rows))
    axes = axes.flatten() 
    
    # Plot each box plot in a separate subplot
    for i, column in enumerate(numeric_columns):
        sns.boxplot(x=data[column], ax=axes[i])
        axes[i].set_title(f"Box Plot of {column}")
        axes[i].set_xlabel("Values")
    
    # Remove any empty subplots if columns do not fill the grid completely
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])
    
    # Adjust layout
    plt.tight_layout()
    plt.show()

# Call the function to plot all boxplots in a grid layout
plot_boxplots_in_grid(spotify)
```

```python
#Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# Function to plot all histograms in one chart with subplots
def plot_histograms_in_grid(data):
    numeric_columns = data.select_dtypes(include=['number']).columns  
    num_columns = len(numeric_columns)  
    
    # Create a grid of subplots (rows and columns)
    cols = 3  
    rows = (num_columns // cols) + (num_columns % cols > 0)  
    
    # Create a figure with subplots
    fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15, 5 * rows))
    axes = axes.flatten()  
    
    # Plot each histogram in a separate subplot
    for i, column in enumerate(numeric_columns):
        sns.histplot(data[column], ax=axes[i], kde= True)
        axes[i].set_title(f"Histogram of {column}")  
        axes[i].set_xlabel("Values")
    
    # Remove any empty subplots if columns do not fill the grid completely
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j]) 
    
    # Adjust layout
    plt.tight_layout() 
    plt.show()  

# Call the function to plot all histograms in a grid layout
plot_histograms_in_grid(spotify)
```

##### Histograms on top (like "artist_count," "released_year") show how often different counts or years appear. For example, most songs seem to have only one artist, and the number of songs released spikes heavily in recent years (after 2000).
##### Columns like "in_spotify_playlists" or "in_apple_playlists" have a right-skewed distribution, where most of the songs are in very few playlists, but there are a few songs in many playlists (rare but extreme values).

##### The histograms for percentage columns like "danceability," "valence," and "energy" are more normally distributed (bell-shaped), showing that most songs have moderate values for these characteristics.

##### Some columns like "instrumentalness" and "liveness" have large numbers of values near zero, indicating that many songs have little or no instrumental or live elements.


```python


# Function to plot the correlation matrix heatmap
def plot_correlation_matrix(data):
    # Select only the numeric columns
    numeric_columns = data.select_dtypes(include=['number'])
    
    # Calculate the correlation matrix
    correlation_matrix = numeric_columns.corr()
    
    # Create the heatmap
    plt.figure(figsize=(10, 8))  # Set the figure size
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    
    # Add title and labels
    plt.title("Correlation Matrix Heatmap")
    plt.show()

# Call the function to plot the correlation matrix for your dataset
plot_correlation_matrix(spotify)
```

##### Perfect correlations on the diagonal (1.00): This means that each variable is perfectly correlated with itself, which is why all diagonal values are 1.00 and show bright red.
##### Strong correlations: For example, "in_spotify_playlists" and "in_deezer_playlists" have a strong positive correlation (0.83), meaning songs in more Spotify playlists tend to also be in more Deezer playlists "in_apple_playlists" and "in_apple_charts" (0.77) also show a strong correlation, meaning songs in Apple playlists are likely to appear in Apple charts.
##### Negative correlations: This means that there are few strong negative correlations, but we do see some slightly negative correlations, like between "released_year" and "in_spotify_playlists" (-0.39). This suggests that older songs are less likely to be in many Spotify playlists.
##### Weaker correlations: Taking a look at some variables like "speechiness" and others related to music features (e.g., "danceability" or "energy") show weak correlations with most other variables, as indicated by values near 0 and blue colors. This means that those features are not strongly related to things like chart placements or playlist counts.
